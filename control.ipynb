```python
import math
def points_angle(p1, p2, p3):
    d12 = dist(p1, p2)
    d23 = dist(p2, p3)
    d31 = dist(p3, p1)
    cos = (-d31**2 + d23**2 + d12**2)/(2*d23*d12)
    return math.acos(cos)*360/2/3.14
def dist(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
def hand_angles(points, l):
    rshoulder = points[l[0]]
    floor = [rshoulder[0], rshoulder[1]+200, rshoulder[2]]
    relbow = points[l[1]]
    rwrist = points[l[2]]
    toret = [None, None]
    toret[0] = points_angle(floor, rshoulder, relbow)
    toret[1] = points_angle(rshoulder, relbow, rwrist)
    probs = (floor[-1]*rshoulder[-1]*relbow[-1], rshoulder[-1]*relbow[-1]*rwrist[-1])
    return toret, probs
def good_hand_angles(points, l, mini=20, maxi=360):
    angles, probs = hand_angles(points, l)
    angles[0] = min(maxi, max(mini, angles[0]*4))
    angles[1] = min(maxi, max(mini, angles[1]*2))
    return angles, probs
def zero_fill(s):
    while len(s) <= 2:
        s = "0" + s
    return s
```
